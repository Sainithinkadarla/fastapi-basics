```python
np.random.seed(0)
m = np.random.randint(10, size = 100000)

def standard_double(array):
    output = np.empty(array.size)
    for i in range(array.size):
        output[i] = array[i] * 2
    return output
```

```python
import timeit
timeit.default_timer()
standard_double(m)
```
